```python
taglist = taglist['letters_new'].str.split(expand=True).add_prefix('position_')    
```

```python
#column_names = ["gene", "pos1", "pos2","pos3","pos4","pos5"]
codebook = pd.DataFrame()
for i,barcode in enumerate(taglist.codebook):
    list_with_gene = [taglist.Gene[i]]
    list_barcode = [int(d) for d in str(barcode)]
    list_with_gene.extend(list_barcode)
    codebook_intermediate = pd.DataFrame([list_with_gene])
    codebook = codebook.append(codebook_intermediate)
```
