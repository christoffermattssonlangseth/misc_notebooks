# evaluate genes and spots per cell 

```python
import pandas as pd
import numpy as np
```

## hca09

```python
cellxgene_mouse = pd.read_csv('/Users/Christoffer/Downloads/HCA09_2_cellxgene.csv')
cellxgene_mouse = cellxgene_mouse.set_index('CellID')
cellxgene_mouse = cellxgene_mouse.drop(['centroidX', 'centroidY'], axis=1)
cellxgene_human = pd.read_csv('/Users/Christoffer/Downloads/HCA09_1_cellxgene.csv')
cellxgene_human = cellxgene_human.set_index('CellID')
cellxgene_human = cellxgene_human.drop(['centroidX', 'centroidY'], axis=1)
```

```python
cellxgene_mouse
```

```python
cellxgene_human
```

### genes per cell

```python
genepercell_mouse = (cellxgene_mouse > 0).astype(int).sum(axis=1)
print('The number of genes per cell in HCA09_mouse is: ' + str(genepercell_mouse.mean()))
```

```python
genepercell_human= (cellxgene_human > 0).astype(int).sum(axis=1)
print('The number of genes per cell in HCA09_human is: ' + str(genepercell_human.mean()))
```

### spots per cell

```python
cellxgene_mouse["sum"] = cellxgene_mouse.sum(axis=1)
cellxgene_human["sum"] = cellxgene_human.sum(axis=1)
```

```python
cellxgene_human
```

```python
print('The number of spots per cell in HCA09_mouse is: ' + str(cellxgene_mouse['sum'].mean()))
```

```python
print('The number of spots per cell in HCA09_human is: ' + str(cellxgene_human['sum'].mean()))
```
