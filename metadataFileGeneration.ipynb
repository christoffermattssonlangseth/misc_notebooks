# get metadata from leica xml file 

this is a first take on how to get metadata information from leica .xml files. this includes the tile number and the x and y position. it is important to note that the current state of the script is very tailored towards a very specific use case. contact christoffer.langseth@scilifelab.se for help on how to set it up. it depends on some packages that make parsing .xml files possible. 

```python
from xml.dom import minidom
import pandas as pd
import numpy as np
import os
```

```python
path = 'N:/TIFF_Export_SL03emb07/'
folders_to_remove = ['else','18D']
output = [dI for dI in os.listdir(path) if os.path.isdir(os.path.join(path,dI))]
output_cleaned = [x for x in output if x not in folders_to_remove]
output_cleaned
```

```python
for i, slide_tag in enumerate(output_cleaned): 
    print(slide_tag)
    for round_num in range(0, 5):
        mydoc = minidom.parse('N:/TIFF_Export_SL03emb07/'+slide_tag+'/MetaData/SL03emb07_'+slide_tag+'_r'+str(round_num+1)+'_Properties.xml')
        tile =[]
        x =[]
        y =[]
        items = mydoc.getElementsByTagName('Tile')
        for elem in items:
            tile.append(int(elem.attributes['FieldX'].value))
            x.append(float(elem.attributes['PosX'].value))
            y.append(float(elem.attributes['PosY'].value))

        dict = {'tile': tile, 'x': x, 'y': y}     
        df = pd.DataFrame(dict) 
        df['tile'] =df.tile+1
        df['x'] =((df.x-np.min(df.x))/.000000321) + 1
        df['y'] =((df.y-np.min(df.y))/.000000321) + 1
        df.to_csv('I:/'+slide_tag+'/preprocessing/2DTiles/tile_coordinates_Base_'+str(round_num+1)+'.csv', index = False)
```
