# assessing base bias

```python
import pandas as pd
import seaborn as sns
```

```python
path_to_file_location = '/Users/Christoffer/Downloads/'
```

```python
barcodes = pd.read_csv(path_to_file_location + 'barcodes.csv')
name_of_column_with_barcodes = 'letters'
```

```python
num_a = barcodes[name_of_column_with_barcodes].str.count('A').sum()
```

```python
num_g =barcodes[name_of_column_with_barcodes].str.count('G').sum()
```

```python
num_c =barcodes[name_of_column_with_barcodes].str.count('C').sum()
```

```python
num_t =barcodes[name_of_column_with_barcodes].str.count('T').sum()
```

```python
counts = [num_a, num_g, num_c, num_t]
```

```python
tag = ['A', 'G', 'C', 'T']
```

```python
ax = sns.barplot(tag,counts)
ax.set(ylabel = "count")
```
